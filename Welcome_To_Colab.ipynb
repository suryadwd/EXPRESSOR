```
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```
data = {
    "Days":["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"],
    "Week1":[27,29,32,33,33,32,34],
    "Week2":[40,23,32,53,33,32,35],
    "Week3":[27,39,35,33,38,36,45],
}

df = pd.DataFrame(data)
df
```

```
plt.plot(df["Days"],df["Week1"])
```

```
w1 = plt.plot(df['Days'],df ["Week1"])
w2 = plt.plot(df['Days'],df ["Week2"])
w3 = plt.plot(df['Days'],df ["Week3"])
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Line plot of week 1 Temperature")
plt.legend(["Week1","Week2","Week3"])
plt.show()
```

```
w1 = plt.plot(df["Days"], df['Week1'],c = 'y',ls='-.')
w2 = plt.plot(df["Days"], df['Week2'], c = 'r',ls="dashed")
w3 = plt.plot(df["Days"], df['Week3'], c = 'g',ls=":")
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Line plot of week 1 Temperature")
plt.legend(["Week1","Week2","Week3"])
plt.show()
```
